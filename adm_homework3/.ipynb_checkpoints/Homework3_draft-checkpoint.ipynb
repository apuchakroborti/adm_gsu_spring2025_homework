```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import rankdata
from itertools import combinations
from math import sqrt
import matplotlib.lines as mlines

# Your accuracy results here
results_dict = {
    "KNN-DTW": {"Dataset1": 0.82, "Dataset2": 0.75, "Dataset3": 0.90, "Dataset4": 0.84, "Dataset5": 0.78,
                "Dataset6": 0.83, "Dataset7": 0.80, "Dataset8": 0.77, "Dataset9": 0.81, "Dataset10": 0.79},
    "TSF":     {"Dataset1": 0.85, "Dataset2": 0.78, "Dataset3": 0.92, "Dataset4": 0.87, "Dataset5": 0.80,
                "Dataset6": 0.84, "Dataset7": 0.82, "Dataset8": 0.79, "Dataset9": 0.83, "Dataset10": 0.80},
    "BOSS":    {"Dataset1": 0.83, "Dataset2": 0.76, "Dataset3": 0.91, "Dataset4": 0.86, "Dataset5": 0.79,
                "Dataset6": 0.82, "Dataset7": 0.81, "Dataset8": 0.78, "Dataset9": 0.82, "Dataset10": 0.78},
    "ROCKET":  {"Dataset1": 0.86, "Dataset2": 0.79, "Dataset3": 0.93, "Dataset4": 0.88, "Dataset5": 0.81,
                "Dataset6": 0.85, "Dataset7": 0.83, "Dataset8": 0.80, "Dataset9": 0.84, "Dataset10": 0.81}
}

classifiers = list(results_dict.keys())
datasets = list(next(iter(results_dict.values())).keys())

# Step 1: Compute ranks (higher accuracy = better rank)
ranks = []
for ds in datasets:
    accs = [-results_dict[clf][ds] for clf in classifiers]
    ranks.append(rankdata(accs, method='average'))

ranks = np.array(ranks)  # shape: (n_datasets, n_classifiers)
avg_ranks = np.mean(ranks, axis=0)

# Step 2: Critical Difference (Nemenyi test)
def nemenyi_cd(k, N, alpha=0.05):
    # Critical values for alpha=0.05 (for k=210); source: Demar (2006)
    q_alpha = {
        2: 1.960, 3: 2.343, 4: 2.569, 5: 2.728,
        6: 2.850, 7: 2.949, 8: 3.031, 9: 3.102, 10: 3.164
    }
    q = q_alpha[k] if k in q_alpha else 3.164  # conservative
    return q * sqrt(k * (k + 1) / (6 * N))

cd = nemenyi_cd(k=len(classifiers), N=len(datasets))
print(f"Critical Difference (CD): {cd:.3f}")

# Step 3: Plot CD diagram
def plot_cd_diagram(avg_ranks, labels, cd, title="CD Diagram"):
    sorted_idx = np.argsort(avg_ranks)
    sorted_ranks = avg_ranks[sorted_idx]
    sorted_labels = [labels[i] for i in sorted_idx]

    fig, ax = plt.subplots(figsize=(10, 2))
    ax.set_title(title)
    ax.set_xlim(min(sorted_ranks) - 0.5, max(sorted_ranks) + 0.5)
    ax.set_ylim(0, 1)
    ax.get_yaxis().set_visible(False)

    # Plot ranks
    for i, (rank, label) in enumerate(zip(sorted_ranks, sorted_labels)):
        ax.plot([rank, rank], [0.1, 0.4], color="black")
        ax.text(rank, 0.5, f"{label}", rotation=45, ha='center', va='bottom')

    # Plot CD bar
    cd_line = mlines.Line2D([sorted_ranks[0], sorted_ranks[0] + cd], [0.8, 0.8], color='black', lw=2)
    ax.add_line(cd_line)
    ax.text(sorted_ranks[0] + cd / 2, 0.85, f"CD = {cd:.2f}", ha='center')

    plt.tight_layout()
    plt.show()

plot_cd_diagram(avg_ranks, classifiers, cd)
```
